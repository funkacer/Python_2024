```python
import csv
from collections import namedtuple
```

```python
data_row = namedtuple('NaceRow', 'sekce oddil skupina trida nazev')
data = []

with open('_NACE_data/nace.csv', 'r') as f:
    reader = csv.reader(f, delimiter='\t', quotechar='"')
    row = next(reader) # zahlavi
    while True:
        #print(row)
        try:
            row = next(reader)
            assert len(row) == 5
            dr = data_row(row[0], row[1], row[2], row[3], row[4])
            # print(dr)
            data.append(dr)
        except Exception as e:
            print(e.__class__)
            print(e)
            break
```

```python
# lepsi verze
data = []

with open('_NACE_data/nace.csv', 'r') as f:
    reader = csv.reader(f, delimiter='\t', quotechar='"')
    row = next(reader) # zahlavi
    data_row = namedtuple('NaceRow', ' '.join(row))
    while True:
        #print(row)
        try:
            row = next(reader)
            assert len(row) == 5
            dr = data_row(*row)
            # print(dr)
            data.append(dr)
        except Exception as e:
            print(e.__class__)
            print(e)
            break
```

```python
# nejlepsi verze
data = []

with open('_NACE_data/nace.csv', 'r') as f:
    reader = csv.reader(f, delimiter='\t', quotechar='"')
    row = next(reader) # zahlavi
    n_column = len(row)
    data_row = namedtuple('NaceRow', ' '.join(row))
    try:
        while (row := next(reader)):
            assert len(row) == n_column
            data.append(data_row(*row))
    except StopIteration as e:
            print("Hotovo! Řádek:", len(data))
```

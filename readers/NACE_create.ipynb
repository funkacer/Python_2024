```python
import csv
import sys
```

```python
rows = []
with open ('_data/Strom.txt', 'r', encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file, delimiter='\t', quotechar='"') #, quoting=csv.QUOTE_NONNUMERIC)
    while True:
        try:
            row = next(reader)
            rows.append(row)
            #print(row)
            #print(len(row))
        except Exception as e:
            print(e)
            #sys.exit()
            break
        #input('Press enter ...')
```

```python
sekce = ''
oddil = ''
skupina = ''
trida = ''
for row in df.sort_values('strom').values:
    #print(row[0], row[1], row[2])
    match row[2].count('/'):
        case 1:
            sekce = row[0]
            print(sekce, row[1])
        case 2:
            oddil = row[0]
            print('\t', oddil, row[1])
        case 3:
            skupina = row[0]
            print('\t', '\t', skupina, row[1])
        case 4:
            trida = row[0]
            print('\t', '\t', '\t', trida, row[1])
        case _:
            pass
```

```python
sekce = ''
oddil = ''
skupina = ''
trida = ''
print('sekce' + '\toddil' + '\tskupina' + '\ttrida' + '\tnazev')
for row in df.sort_values('strom').values[:]:
    #print(row[0], row[1], row[2])
    match row[2].count('/'):
        case 1:
            sekce = row[0]
            print(sekce + '\t' + '\t' + '\t' + '\t' + row[1])
        case 2:
            oddil = row[0]
            print(sekce +f'\t{oddil}' + '\t' + '\t' + '\t' + row[1])
        case 3:
            skupina = row[0]
            print(sekce + f'\t{oddil}' + f'\t{skupina}' + '\t' + '\t' + row[1])
        case 4:
            trida = row[0]
            print(sekce + f'\t{oddil}' + f'\t{skupina}' + f'\t{trida}' + '\t' + row[1])
        case _:
            pass
```

```python
sekce = ''
oddil = ''
skupina = ''
trida = ''
with open('nace.csv', 'w') as f:
    #print('Filename:', filename, file=f)  # Python 3.x
    print('sekce' + '\toddil' + '\tskupina' + '\ttrida' + '\tnazev', file=f)
    for row in df.sort_values('strom').values[:]:
        #print(row[0], row[1], row[2])
        match row[2].count('/'):
            case 1:
                sekce = row[0]
                print(sekce + '\t' + '\t' + '\t' + '\t' + row[1], file=f)
            case 2:
                oddil = row[0]
                print(sekce + f'\t{oddil}' + '\t' + '\t' + '\t' + row[1], file=f)
            case 3:
                skupina = row[0]
                print(sekce + f'\t{oddil}' + f'\t{skupina}' + '\t' + '\t' + row[1], file=f)
            case 4:
                trida = row[0]
                print(sekce + f'\t{oddil}' + f'\t{skupina}' + f'\t{trida}' + '\t' + row[1], file=f)
            case _:
                pass
```

```python
import csv
import numpy as np
import pandas as pd
```

```python
def read_csv(file):
    with open (file=file, mode='r', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';', quotechar='"')
        header = next(csv_reader)
        for i, column in enumerate(header):
            header[i] = column.strip().replace('\ufeff', '')
        n_columns = len(header)
        n_skipped_rows = 0
        skip_row = False
        rows = []
        for i, row in enumerate(csv_reader):
            print(', '.join(row))
            assert len(row) == n_columns, 'Počet sloupců je divný'
            for j, column in enumerate(header):
                if j == 0:    #radky
                    if row[j] != '':
                        rows.append([row[j]])
                        skip_row = False
                    else:
                        skip_row = True
                        n_skipped_rows += 1
                elif not skip_row:    #cisla
                    try:
                        cislo = float(row[j].strip().replace(' ', '').replace(',', '.'))
                    except Exception as e:
                        cislo = np.nan
                    rows[i].append(cislo)
    assert csv_reader.line_num - 1 - n_skipped_rows == len(rows), 'Počet řádků je divný'
    return header, rows
```

```python
file = '_CR_242_PLS_M8_reader_data/CR_242_PLS_M8_old.csv'
```

```python
columns, data = read_csv(file)
```

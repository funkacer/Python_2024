```python
import csv
import sys
```

```python
rows = []
with open ('_data/Strom.txt', 'r', encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file, delimiter='\t', quotechar='"') #, quoting=csv.QUOTE_NONNUMERIC)
    while True:
        try:
            row = next(reader)
            rows.append(row)
            #print(row)
            #print(len(row))
        except Exception as e:
            print(e)
            #sys.exit()
            break
        #input('Press enter ...')
```

```python
hlavni_trida = ''
trida = ''
skupina = ''
podskupina = ''
kategorie = ''
for row in df.sort_values('strom').values:
    #print(row[0], row[1], row[2])
    match row[2].count('/'):
        case 1:
            hlavni_trida = row[0]
            print(hlavni_trida, row[1])
        case 2:
            trida = row[0]
            print('\t', trida, row[1])
        case 3:
            skupina = row[0]
            print('\t', '\t', skupina, row[1])
        case 4:
            if (isinstance(row[2], str) and row[2][-1] == '/'):
                podskupina = row[0]
                print('\t', '\t', '\t', podskupina, row[1])
            else:
                kategorie = row[0]
                print('\t', '\t', '\t', '\t', kategorie, row[1])
        case _:
            pass
```

```python
hlavni_trida = ''
trida = ''
skupina = ''
podskupina = ''
kategorie = ''
with open('isco.csv', 'w') as f:
    #print('Filename:', filename, file=f)  # Python 3.x
    print('hlavni_trida' + '\ttrida' + '\tskupina' + '\tpodskupina' + '\tkategorie' + '\tnazev', file=f)
    for row in df.sort_values('strom').values[:]:
        #print(row[0], row[1], row[2])
        match row[2].count('/'):
            case 1:
                hlavni_trida = row[0]
                print(hlavni_trida + '\t' + '\t' + '\t' + '\t' + f'\t{row[1]}', file=f)
            case 2:
                trida = row[0]
                print(hlavni_trida + f'\t{trida}' + '\t' + '\t' + '\t' + f'\t{row[1]}', file=f)
            case 3:
                skupina = row[0]
                print(hlavni_trida + f'\t{trida}' + f'\t{skupina}' + '\t' + '\t' + f'\t{row[1]}', file=f)
            case 4:
                if (isinstance(row[2], str) and row[2][-1] == '/'):
                    podskupina = row[0]
                    print(hlavni_trida + f'\t{trida}' + f'\t{skupina}' + f'\t{podskupina}' + '\t' + f'\t{row[1]}', file=f)
                else:
                    kategorie = row[0]
                    print(hlavni_trida + f'\t{trida}' + f'\t{skupina}' + f'\t{podskupina}' + f'\t{kategorie}' + f'\t{row[1]}', file=f)
            case _:
                pass
```

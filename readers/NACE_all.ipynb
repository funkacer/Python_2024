```python
import csv
import sys
```

```python
rows = []
with open ('_NACE_data/Strom.txt', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t', quotechar='"') #, quoting=csv.QUOTE_NONNUMERIC)
    while True:
        try:
            row = next(reader)
            rows.append(row)
            #print(row)
            #print(len(row))
        except Exception as e:
            print(e)
            #sys.exit()
            break
        #input('Press enter ...')
```

# Nace rows jako named tuple

```python
from collections import namedtuple

nr = namedtuple("NaceRow", "sekce oddil skupina trida nazev")

nace_rows = []

sekce = ''
oddil = ''
skupina = ''
trida = ''
#print('sekce' + '\toddil' + '\tskupina' + '\ttrida' + '\tnazev')
nace_row = nr('sekce', 'oddil', 'skupina', 'trida', 'nazev')
nace_rows.append(nace_row)
for row in df.sort_values('strom').values[:]:
    #print(row[0], row[1], row[2])
    match row[2].count('/'):
        case 1:
            sekce = row[0]
            nace_row = nr(sekce, None, None, None, row[1])
            #print(sekce + '\t' + '\t' + '\t' + '\t' + row[1])
        case 2:
            oddil = row[0]
            nace_row = nr(sekce, oddil, None, None, row[1])
            #print(sekce + f'\t{oddil}' + '\t' + '\t' + '\t' + row[1])
        case 3:
            skupina = row[0]
            nace_row = nr(sekce, oddil, skupina, None, row[1])
            #print(sekce + f'\t{oddil}' + f'\t{skupina}' + '\t' + '\t' + row[1])
        case 4:
            trida = row[0]
            nace_row = nr(sekce, oddil, skupina, trida, row[1])
            #print(sekce + f'\t{oddil}' + f'\t{skupina}' + f'\t{trida}' + '\t' + row[1])
        case _:
            pass
    nace_rows.append(nace_row)
```

# Nace rows do csv

```python
import csv
with open('nace_excel.csv', 'w', newline='', encoding='utf-8') as csvfile:
    #writer = csv.writer(csvfile, delimiter=';', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    #writer.writerow(['Spam'] * 5 + ['Baked Beans'])
    #writer.writerow(['Spam', 'Lovely Spam', 'Wonderful Spam'])
    fieldnames = nace_rows[0]._asdict()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter=';',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL) #csv.QUOTE_MINIMAL dialect='excel'
    writer.writeheader()
    #writer.writerow({'first_name': 'Baked', 'last_name': 'Beans'})

    for nace_row in nace_rows[1:]:
        writer.writerow(nace_row._asdict())
```

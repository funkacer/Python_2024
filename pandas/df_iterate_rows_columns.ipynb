```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.rand(100).reshape(10,10), columns = 'a b c d e f g h i j'.split())
```

```python
df
```

```python
df[['a', 'b'] + df.columns[-5:].to_list()]
```

```python
# po řádcích
for index in df.index:
    print(index)
    print(df.loc[index])
```

```python
# po řádcích 2
for index in df.index:
    print(index)
    for column_value in df.loc[index].items():
        print(column_value)
```

```python
# po sloupcích
for column, rows in df.items():
    print(column)
    print(rows)
```

```python
# pos sloupcích 2
for column, rows in df.items():
    print(column)
    for index, row in rows.items():
        print(index, row)
```

# df1 groupby count() vs size()

```python
df1.groupby(by = ['A', 'B'])['a'].count()
```

```python
# to samé, jen chybí název
df1.groupby(by = ['A', 'B']).size()
```

```python
df1.groupby(by = ['A', 'B'])['a'].count().to_frame()
```

```python
df1.groupby(by = ['A', 'B'])['a'].count().to_frame(name = 'count')
```

```python
df1.groupby(by = ['A', 'B'])['a'].count().reset_index(name = 'count')
```

```python
# nejjednodušší přes size:
df1.groupby(['A', 'B']).size().to_frame(name='count')
```

```python
df1.groupby(by = ['A', 'B']).agg(func = {'a':['sum', 'mean', 'median']})
```

# df3 multiindex přes řádky i sloupce

```python
df3 = df1.groupby(by = ['A', 'B']).agg(func = {'a':['sum', 'mean', 'median']})
df3.loc[('A', 'B'), ('a', 'mean')]
```

# agregace bez multiindexů

```python
gb = df1.groupby(['A', 'B'])
counts = gb.size().to_frame(name='count')
counts = (counts
 .join(gb.agg({'a': 'sum'}).rename(columns={'a': 'a_sum'}))
 .join(gb.agg({'a': 'mean'}).rename(columns={'a': 'a_mean'}))
 .join(gb.agg({'a': 'median'}).rename(columns={'a': 'a_median'}))
 .reset_index()
)
```

```python
counts
```

```python
df1.groupby(['A', 'B']).agg({'a':'sum'})
```

```python
# tohle udělá multiindex
df1.groupby(['A', 'B']).agg({'a':['sum']})
```

```python
# fígle
df4 = df1[['A', 'B', 'a']].astype(str)
df4['a'] = df4['a'].str[:5].str.pad(width = 6, side='left', fillchar='0')
df4.groupby(['A', 'B']).agg({'a':', '.join})
```

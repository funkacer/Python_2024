```python
random_size = 9
```

```python
startday = datetime.date(2024, 1, 1)
endday = datetime.date(2024, 6, 30)
np.busday_count(begindates=startday, enddates=endday, weekmask='1111100', holidays=[])
```

```python
random_dates = []
for i in range(random_size):
    bimonthly_days = np.arange(0, 60)
    base_date = np.datetime64('2024-05-01')
    random_date = base_date + np.random.choice(bimonthly_days)
    random_dates.append(random_date)
random_dates
```

```python
df = pd.DataFrame(random_dates, columns=['konecep'])
df
```

```python
# eviddny random 1-182
eviddny = np.random.randint(low=1, high=100, size=random_size)
eviddny
```

```python
# udělej named series eviddny
eviddny_ps = pd.Series(eviddny, dtype=np.int64, name='eviddny')
eviddny_ps
```

```python
# prida prazdny sloupec
df_nechci = pd.concat([df, pd.Series()], ignore_index=True)
df_nechci
```

```python
# add series to df
# DataFrame.insert(loc, column, value, allow_duplicates=<no_default>)
df.insert(loc = 0, column = 'konecep_old', value = pd.Series())
df
```

```python
# add series to df
# DataFrame.insert(loc, column, value, allow_duplicates=<no_default>)
df = df.join(eviddny_ps)
df
```

```python
#konecep bylo puvodne cislo, udelam cislo
df['konecep_old'] = df['konecep']
df
```

```python
df.dtypes
```

```python
#konecep bylo puvodne cislo, udelam cislo
df['konecep'] = df['konecep_old'].astype(str).str.replace('-','').astype(int)
df
```

```python
df.dtypes
```

```python
# přidat řádek s celým rozsahem
df = pd.concat([df, pd.DataFrame({'konecep':[20240630], 'eviddny':[182]})], ignore_index=True, axis=0)
df
```

```python
# konecep jako datum konecday, z toho startday
#df['konecdays'] = pd.to_datetime(df['konecep'], format='%Y%m%d')
#df.drop(columns=['konecdays'], inplace=True)
df['konecday'] = pd.to_datetime(df['konecep'], format='%Y%m%d')
df
```

```python
df['startday'] = df['konecday'] -  pd.to_timedelta(df['eviddny']-1, unit='d')
df
```

```python
startdays = [d.date() for d in df['startday'].to_list()]
enddays = [d.date() for d in df['konecday'].to_list()]
df['businesdays'] = np.busday_count(begindates=startdays, enddates=enddays, weekmask='1111100', holidays=[])
df
```

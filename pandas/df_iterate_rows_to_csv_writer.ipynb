```python
import pandas as pd
import numpy as np
```

# df iterate rows and columns

```python
df = pd.DataFrame(np.random.rand(100).reshape(10,10), columns = 'a b c d e f g h i j'.split())
```

```python
df
```

# df1 - groupby sum()

```python
keys = np.array([
        ['A', 'B'],
        ['A', 'B'],
        ['A', 'B'],
        ['A', 'B"B'],
        ['C', 'D'],
        ['C', 'D'],
        ['C', 'DD'],
        ['E', 'F'],
        ['E', 'FF'],
        ['G', 'H'] 
        ])
```

```python
df1 = pd.DataFrame(keys, columns = ['A', 'B']).join(df)
df1
```

```python
export_df = df1.groupby(by = ['A', 'B']).sum()
export_df
```

```python
import csv

export_df = df1.groupby(by = ['A', 'B']).sum()

with open(file = 'export_df_writer.csv', mode = 'w', newline='', encoding='utf-8',) as csvfile:
    fieldnames = export_df.index.names + export_df.columns.to_list()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter='\t', quotechar='"', quoting=csv.QUOTE_NONNUMERIC) #quoting=csv.QUOTE_MINIMAL)
    writer.writeheader()
    for index in export_df.index:
        writer.writerow({k:v for k,v in [i for i in zip(export_df.index.names, index)] + [i for i in export_df.loc[index].items()]})
```

```python
import csv

with open(file = 'export_df_writer.csv', mode = 'r', encoding='utf-8',) as csvfile:
    reader = csv.DictReader(csvfile, delimiter='\t', quotechar='"')
    fieldnames = reader.fieldnames
    print('Fieldnames:', fieldnames, sep='\n')
    print('Rows:')
    line = next(reader)
    while True:
        print(line)
        try:
            line = next(reader)
        except Exception as e:
            print(e)
            break
```

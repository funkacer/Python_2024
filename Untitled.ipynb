```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi*2, np.pi*2, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
plt.plot(x, y)
plt.show()
```

```python
y = np.cos(x)
```

```python
y
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
plt.plot(x, y)
plt.show()
```

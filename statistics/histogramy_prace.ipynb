```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
y = np.random.normal(0, 30, 1000)
```

```python
bins = 11
hist, bin_edges = np.histogram(y, bins, density = True)
bin_edges
```

```python
np.diff(bin_edges)
```

```python
bin_centers = bin_edges[:-1] + np.diff(bin_edges)/2
bin_centers
```

```python
hist
```

```python
mu = np.mean(y)
variance = np.var(y)
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.bar(bin_centers, hist, color = "g", alpha = 0.5, width = 6*sigma/bins)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

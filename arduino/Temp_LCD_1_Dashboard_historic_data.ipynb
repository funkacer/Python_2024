```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#folder = './_data/historic-data-20241228T184051Z'
folder = './_data/historic-data-20241230T173547Z'
variables = ['humi', 'tempC', 'tempF']
datum = 'datum'
```

```python
def get_data(file):
    data = []
    with open (file=file, mode='r', encoding='utf-8') as f:
        reader = csv.reader(f, delimiter=',')
        for row in reader:
            print(','.join(row))
            data.append(row)
    return data
```

```python
data = []
for variable in variables:
    data.append(get_data(folder + '/Temp LCD 1-' + variable + '.csv'))
```

```python
data
```

```python
dfs = []
for i, d in enumerate(data):
    dfs.append(pd.DataFrame(d[1:], columns = [datum, variables[i]]))
```

```python
import functools as ft
#df = ft.reduce(lambda left, right: pd.merge(left, right, on=datum), dfs) #udela inner join
df = ft.reduce(lambda left, right: pd.merge(left, right, on=datum, how='outer'), dfs)
```

```python
df.info()
```

```python
df[variables] = df[variables].apply(pd.to_numeric)
df
```

```python
# pd.to_datetime('2018-10-26 12:00:00.0000000011', format='%Y-%m-%d %H:%M:%S.%f')
df[datum] = df[datum].apply(pd.to_datetime, format='%Y-%m-%dT%H:%M:%S.%fZ')
df
```

```python
df.describe()
```

```python
plt.plot(df.sort_values(datum)[datum], df.sort_values(datum)[variables], label=variables)
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```
